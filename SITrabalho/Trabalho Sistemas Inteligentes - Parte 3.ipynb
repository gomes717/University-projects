<h2> Importando as bibliotecas necessárias </h2>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

from pylab import *
from matplotlib import pylab
import matplotlib.pyplot as plt
```

<h2> Abertura do Arquivo</h2>

```python
df = pd.read_excel("dados_compilados.xlsx")
```

<h2> Preparação dos dados </h2>

```python
df_val = df[['new_cases', 'reproduction_rate', 'people_fully_vaccinated', 'ocupacao']]
df_resul = df['ocupacao_in_7_days']
x = df_val.values.tolist()
y = df_resul
```

<h2> Criação dos dados de treino e de teste </h2>

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, random_state=42)
```

<h2> Normalização dos dados </h2>

```python
scaler = StandardScaler()
scaler.fit(x_treino)
x_treino = scaler.transform(x_treino)
x_teste = scaler.transform(x_teste)
```

<h2> Treinamento da Rede neural</h2>

```python
mlp = MLPRegressor(hidden_layer_sizes=(40, 30, 20), max_iter=5000, activation='relu', verbose=True)
a = mlp.fit(x_treino, y_treino)
```

<h2> Testes </h2>

```python
labels = mlp.predict(x_teste)
error = y_teste - labels
diferenca = np.absolute(error)
porcentagem_em_relacao_a_teste = (diferenca * 100)/y_teste
```

<h2> Plotagem dos gráficos </h2>

```python
plt.plot([*range(0, len(labels), 1)], labels, '-', color='black', label='Previsões')
plt.plot([*range(0, len(y_teste), 1)], y_teste, '-', color='red', label='Reais')
plt.grid(True)
legend()
```

```python
plt.plot([*range(0, len(labels), 1)], porcentagem_em_relacao_a_teste, color='blue', label='erro')
plt.grid(True)
pylab.title('Erro', size = 19)
print("Erro médio: ", np.mean(porcentagem_em_relacao_a_teste))
```

```python
plt.plot([*range(0, len(a.loss_curve_), 1)], a.loss_curve_, '-', color='red', label='Reais')
plt.grid(True)
```
